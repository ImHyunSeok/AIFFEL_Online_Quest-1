```python
# 메뉴와 가격을 리스트 형식으로 작성
menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
price = [2000, 3000, 3000, 2500, 2500, 3000]

# 키오스크 클래스 생성
class Kiosk:
    # Kiosk 클래스가 실행될때 메뉴와 가격을 불러온다.    
    def __init__(self):
        self.menu = menu
        self.price = price
    # 메뉴 목록과 가격 목록을 출력한다.     
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i+1, self.menu[i], ' : ', self.price[i], '원')
    # 주문     
    def menu_select(self):
        #주문한 품목과 그에 해당하는 가격을 빈리스트로 지정         
        self.order_menu = []  
        self.order_price = []  
        #음료 주문을 위한 변수 n을 초기값 0으로 설정
        n = 0
        #음료 번호가 유효하지 않은 경우 while loop을 계속 돌린다.        
        while n < 1 or len(menu) < n:
            #음료 번호를 입력
            n = int(input("음료를 번호를 입력하세요 : "))  

            #만약 음료 번호가 유효한 경우 위의 주문가격 리스트를 업데이트
            if 1 <= n & n <= len(menu):
                self.order_price.append(self.price[n-1])
                #주문 가격의 합계를 별도 변수로 지정
                self.price_sum = sum(self.order_price)  
            #유효하지 않은 경우 없는 메뉴입니다. 다시 주문해주세요 문구를 출력
            else:  
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        #음료 온도 설정을 위한 변수 t의 초기값을 0으로 설정
        t = 0
        #음료 온도 설정 번호가 유효하지 않은 경우 while loop을 계속 실행
        while t != 1 and t != 2:
            #Hot 음료는 1 , ICE는 2를 입력한다.
            t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
            if t == 1:    
                self.temp = 'HOT'
            elif t ==2:
                self.temp ='ICE'
            else:    
                print("1과 2 중 하나를 입력하세요.\n")
            
        # 주문 품목을 '온도' '품명'으로 리스트로 저장
        self.order_menu.append(self.temp + ' ' + self.menu[n-1])
        #주문한 품목을 '온도' '품명' : '가격' '원' 형식으로 출력
        print(self.temp, self.menu[n-1], ' : ', self.price[n-1], '원')  

        #추가 주문을 위한 while loop 형성
        while n != 0:
            #줄바꿈
            print()
            #추가 주문을 원할 시 음료 번호를 입력 지불을 원할 시 0을 입력
            n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))
            #주문 번호가 유효한 경우 온도 while loop을 실행한다.
            if 1 <= n & n <= len(menu): 
                t = 0
                while t != 1 and t != 2:  
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:    
                        self.temp = 'HOT'
                    elif t ==2:
                        self.temp ='ICE' 
                    else:    
                        print("1과 2 중 하나를 입력하세요.\n") 
                #주가 주문가격 리스트 및 합계, 주문 품목 리스트를 업데이트
                self.order_price.append(self.price[n-1])  
                self.price_sum = sum(self.order_price)
                self.order_menu.append(self.temp + ' ' + self.menu[n-1])  
                #추가 주문 품목과 가격 합계를 출력
                print('추가 주문 음료', self.temp, self.menu[n-1], ':', self.price[n-1], '원\n', '합계 : ', self.price_sum, '원')
            #주문 번호가 유효하지 않은 경우
            else :
                #지불을 입력한 경우 주문이 완료되었습니다 출력 아닌 경우 다시 위 쪽 없는 메뉴 출력 후 while loop 반복
                if n == 0 :  
                    print("주문이 완료되었습니다.")
                    print(self.order_menu, self.order_price)  
                else :
                    print("없는 메뉴입니다. 다시 주문해 주세요.")
    # 지불 방식 선택    
    def pay(self):
        #합계 금액 출력
        print('지불할 금액은',self.price_sum,'원입니다.')
        #카드 혹은 현금 선택 메뉴
        n = int(input("""현금('cash' press 1).\n카드('card' press 2)\n"""))
        #다른 번호를 선택한 경우 while loop 반복
        while n != 1 or n != 2:
            #현금, 카드 선택시 출력 문구를 설정, 잘못 누른 경우 다시 결제 유도후 while loop 반복
            if n == 1:
                print("'직원을 호출하겠습니다.'")
                break
            elif n == 2:
                print("'IC칩 방향에 맞게 카드를 꽂아주세요.'")
                break
            else:
                print('다시 결제를 시도해 주세요.')
    #주문서 출력            
    def table(self): 
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i])

        print('합계 금액 :', self.price_sum)

        for i in range(5):
            print('|' + ' ' * 31+ '|')
        print('⟝' + '-' * 30 + '⟞') 

a = Kiosk() 
a.menu_print()  
a.menu_select() 
a.pay()  
a.table()  
```
